## Using Fuzzywuzzy library over here

```python
import csv
import pandas as pd
from fuzzywuzzy import fuzz

# Function to read data from CSV file
def read_data_from_csv(file_path):
    data = []
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data

source_csv_path = '../data/Final/Base.csv'

reference_csv_path = '../data/Final/Reference.csv'

source_data = read_data_from_csv(source_csv_path)
reference_data = read_data_from_csv(reference_csv_path)

def fuzzy_match_records(source_record, reference_records, threshold=70):
    matches = []
    for ref_record in reference_records:
        name_similarity = fuzz.token_sort_ratio(source_record["FullNameEn"], ref_record["FullNameEn"])
        if name_similarity >= threshold:
            matches.append({
                'BaseData': source_record,
                'ReferenceData': ref_record,
                'NameSimilarity': name_similarity
            })
    return matches

matches = []
for source_record in source_data:
    matches.extend(fuzzy_match_records(source_record, reference_data))

matched_df = pd.DataFrame(matches)

# Save matched results to CSV
matched_df.to_csv('matched_results_fuzzywuzzy.csv', index=False)

# Print matched records
for index, match in matched_df.iterrows():
    print("Base Data:")
    print(match['BaseData'])
    print("Reference Data:")
    print(match['ReferenceData'])
    print("Name Similarity:", match['NameSimilarity'])
    print()
```

## Ratcliff/Obershelp Pattern Recognition algorithm (‘SequenceMatcher’)


```python
import csv
import pandas as pd
from difflib import SequenceMatcher

# Function to read data from CSV file
def read_data_from_csv(file_path):
    data = []
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data

source_csv_path = '../data/Final/Base.csv'

reference_csv_path = '../data/Final/Reference.csv'

source_data = read_data_from_csv(source_csv_path)
reference_data = read_data_from_csv(reference_csv_path)

def sequence_match_records(source_record, reference_records, threshold=0.6):
    matches = []
    for ref_record in reference_records:
        name_similarity = SequenceMatcher(None, source_record["FullNameEn"], ref_record["FullNameEn"]).ratio()
        if name_similarity >= threshold:
            matches.append({
                'BaseData': source_record,
                'ReferenceData': ref_record,
                'NameSimilarity': name_similarity
            })
    return matches

matches = []
for source_record in source_data:
    matches.extend(sequence_match_records(source_record, reference_data))

matched_df = pd.DataFrame(matches)

# Save matched results to CSV
matched_df.to_csv('matched_results_sequence_matcher.csv', index=False)

# Print matched records
for index, match in matched_df.iterrows():
    print("Base Data:")
    print(match['BaseData'])
    print("Reference Data:")
    print(match['ReferenceData'])
    print("Name Similarity:", match['NameSimilarity'])
    print()
```

##  Levenshtein algorithm (‘FuzzyWuzzy’)


```python
import csv
import pandas as pd
from fuzzywuzzy import fuzz

# Function to read data from CSV file
def read_data_from_csv(file_path):
    data = []
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data

source_csv_path = '../data/Final/Base.csv'

reference_csv_path = '../data/Final/Reference.csv'

# Read data from CSV files
source_data = read_data_from_csv(source_csv_path)
reference_data = read_data_from_csv(reference_csv_path)

# Function to perform fuzzy matching using Levenshtein algorithm
def levenshtein_match_records(source_record, reference_records, threshold=60):
    matches = []
    for ref_record in reference_records:
        name_similarity = fuzz.ratio(source_record["FullNameEn"], ref_record["FullNameEn"])
        if name_similarity >= threshold:
            matches.append({
                'BaseData': source_record,
                'ReferenceData': ref_record,
                'NameSimilarity': name_similarity
            })
    return matches

# Perform fuzzy matching using Levenshtein algorithm
matches = []
for source_record in source_data:
    matches.extend(levenshtein_match_records(source_record, reference_data))

# Convert matched results to pandas DataFrame
matched_df = pd.DataFrame(matches)

# Save matched results to CSV
matched_df.to_csv('matched_results_levenshtein.csv', index=False)

# Print matched records
for index, match in matched_df.iterrows():
    print("Base Data:")
    print(match['BaseData'])
    print("Reference Data:")
    print(match['ReferenceData'])
    print("Name Similarity:", match['NameSimilarity'])
    print()
```

## Jaccard Algorithm

```python
import csv
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Function to read data from CSV file
def read_data_from_csv(file_path):
    data = []
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data

source_csv_path = '../data/Final/Base.csv'

reference_csv_path = '../data/Final/Reference.csv'

# Read data from CSV files
source_data = read_data_from_csv(source_csv_path)
reference_data = read_data_from_csv(reference_csv_path)

# Function to perform fuzzy matching using Jaccard similarity
def jaccard_match_records(source_record, reference_records, threshold=0.4):
    matches = []
    source_name_tokens = set(source_record["FullNameEn"].split())
    for ref_record in reference_records:
        ref_name_tokens = set(ref_record["FullNameEn"].split())
        jaccard_similarity = len(source_name_tokens.intersection(ref_name_tokens)) / len(source_name_tokens.union(ref_name_tokens))
        if jaccard_similarity >= threshold:
            matches.append({
                'BaseData': source_record,
                'ReferenceData': ref_record,
                'NameSimilarity': jaccard_similarity
            })
    return matches

# Perform fuzzy matching using Jaccard similarity
matches = []
for source_record in source_data:
    matches.extend(jaccard_match_records(source_record, reference_data))

# Convert matched results to pandas DataFrame
matched_df = pd.DataFrame(matches)

# Save matched results to CSV
matched_df.to_csv('matched_results_jaccard.csv', index=False)

# Print matched records
for index, match in matched_df.iterrows():
    print("Base Data:")
    print(match['BaseData'])
    print("Reference Data:")
    print(match['ReferenceData'])
    print("Name Similarity:", match['NameSimilarity'])
    print()
```
