## Check same mobile number tagged to multiple names

```python
from fuzzywuzzy import fuzz
import pandas as pd

# Assuming you have a DataFrame called df with columns 'FullNameEn' and 'mobile_number'

# Function to check if two names are similar
def are_names_similar(name1, name2):
    # You can adjust the threshold according to your requirements
    similarity_ratio = fuzz.token_sort_ratio(name1, name2)
    return similarity_ratio > 59  # Adjust threshold as needed

# Grouping by mobile number and checking for duplicate names
grouped = df.groupby('mobile_number')['FullNameEn'].apply(list)

# Initialize a list to store the results
duplicates = []

# Iterate through each group
for mobile, names in grouped.iteritems():
    # If there's only one name associated with the mobile number, skip
    if len(names) <= 1:
        continue
    
    # Iterate through pairs of names to check for similarity
    for i in range(len(names)):
        for j in range(i + 1, len(names)):
            if are_names_similar(names[i], names[j]):
                # If the names are similar, consider it a duplicate
                duplicates.append({'mobile_number': mobile, 'name1': names[i], 'name2': names[j]})

# Convert the list of duplicates to a DataFrame
duplicates_df = pd.DataFrame(duplicates)
duplicates_df
```

## On name, apply fuzzy logic and label as perfect duplicate and potential duplicate

```python
import csv
import pandas as pd
from fuzzywuzzy import fuzz

# Function to read data from CSV file
def read_data_from_csv(file_path):
    data = []
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data


source_csv_path = 'Family_Member.csv'
reference_csv_path = '15May2024_ReferenceData.csv'

source_data = read_data_from_csv(source_csv_path)
reference_data = read_data_from_csv(reference_csv_path)

def fuzzy_match_records(source_record, reference_records, threshold=90):
    matches = []
    for ref_record in reference_records:
        name_similarity = fuzz.token_sort_ratio(source_record["FullNameEn"], ref_record["FullNameEn"])
        if name_similarity >= threshold:
            matches.append({
                'BaseData': source_record,
                'ReferenceData': ref_record,
                'NameSimilarity': name_similarity
            })
    return matches

matches = []
for source_record in source_data:
    matches.extend(fuzzy_match_records(source_record, reference_data))

matched_df = pd.DataFrame(matches)

# Add indicator column
matched_df['DuplicateType'] = matched_df.apply(lambda x: 'Perfect Duplicate' if x['BaseData']['FullNameEn'] == x['ReferenceData']['FullNameEn'] else 'Potential Duplicate', axis=1)

# Save matched results to CSV
matched_df.to_csv('matched_results_fuzzywuzzy.csv', index=False)

# Print matched records
for index, match in matched_df.iterrows():
    print("Base Data:")
    print(match['BaseData'])
    print("Reference Data:")
    print(match['ReferenceData'])
    print("Name Similarity:", match['NameSimilarity'])
    print("Duplicate Type:", match['DuplicateType'])
    print()
```
