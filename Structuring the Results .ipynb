```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
import numpy as np
pd.set_option('display.float_format', '{:.6f}'.format)
pd.set_option('display.float_format', lambda x: "{:.6f}".format(x) if not pd.isna(x) else "NaN")
```

```python
data=pd.read_csv("matched_results_jaccard.csv")
```

```python
data
```

```python
import ast
```

```python
data['BaseData'] = data['BaseData'].apply(ast.literal_eval)
data['ReferenceData'] = data['ReferenceData'].apply(ast.literal_eval)
```

```python
base_data_df = pd.json_normalize(data['BaseData'])
reference_data_df = pd.json_normalize(data['ReferenceData'])
```

```python
result_df = pd.concat([base_data_df, reference_data_df, data['NameSimilarity']], axis=1)
```

```python
result_df
```

```python
base_data_columns = [f'BaseData_{col}' for col in base_data_df.columns]
reference_data_columns = [f'ReferenceData_{col}' for col in reference_data_df.columns]
result_df.columns = base_data_columns + reference_data_columns + ['NameSimilarity']
```
